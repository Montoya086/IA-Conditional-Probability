```python
import random
import matplotlib.pyplot as plt
```

## Ejercicio 1

### Descripci√≥n: Mientras se obtenga un 1 o un 6, el dado es lanzado nuevamente. En caso contrario, el experimento termina.

```python
# Define a function to simulate the dice experiment
def dice_simulator():
    count = 0
    while True:
        roll = random.randint(1, 6)
        count += 1
        if roll not in [1, 6]:
            break
    return count

# Define a function to simulate the dice experiment N times and plot the probability mass function
def simulate_and_plot(N):
    results = [dice_simulator() for _ in range(N)]
    
    # Calculate the probability mass function
    pmf = [results.count(i)/N for i in range(1, max(results)+1)]
    
    # Plot
    plt.bar(range(1, max(results)+1), pmf)
    plt.title(f"Probability Mass Function for N = {N}")
    plt.xlabel("Number of dice rolls")
    plt.ylabel("Probability")
    plt.show()

# Simulate the dice experiment for different values of N
for N in [10, 100, 1000, 10000]:
    simulate_and_plot(N)
```

## Ejercicio 2

### Descripci√≥n: Se lanzan 2 dados regulares, y se obtiene el valor de la suma de ambos D1+D2.

```python
# Define a function to simulate the dice
def dice_simulator():
    sum = 0
    for _ in range(2):
        roll = random.randint(1, 6)
        sum += roll
    return sum

# Define a function to simulate the dice experiment N times
def simulate(N):
    results = [dice_simulator() for _ in range(N)]
    
    # Calculate the average
    average = sum(results)/N
    return average

def plot_averages(N):
    averages = [simulate(N) for _ in range(1000)]
    
    # Plot
    plt.hist(averages, bins=30)
    plt.title(f"Histogram of averages for N = {N}")
    plt.xlabel("Average")
    plt.ylabel("Frequency")
    plt.show()

for N in [5, 10, 50, 100, 200]:
    plot_averages(N)
```
